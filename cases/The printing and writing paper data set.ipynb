# Comparative case: The printing and writing paper data set

```python
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

import skfore

import warnings
warnings.filterwarnings('ignore')
```

Load data

```python
from skfore.datasets import load_paper
```

```python
ts = load_paper()
tse = ts[0:100]
tsp = ts[100:]
```

```python
import sklearn
from IPython.display import HTML, display
import tabulate
```

### Modelo AMNFIS

```python
warnings.filterwarnings('ignore')
from skfore.models.AMNFIS import AMNFIS
```

```python
model = AMNFIS(p = 4, c = 4)
model.fit(tse)
forecast = model.predict(ts = tse, periods = 20)
MSE_E = model.calc_error(ts = tse)
MAD_E = model.calc_error(ts = tse, error_function = sklearn.metrics.mean_absolute_error)
MSE_P = model.calc_error(ts = tsp)
MAD_P = model.calc_error(ts = tsp, error_function = sklearn.metrics.mean_absolute_error)
model.plot(tse, forecast)
```

```python
table = [["Error","Entrenamiento","Prueba"],
         ["MSE",MSE_E,MSE_P],
         ["MAD",MAD_E,MAD_P]]

display(HTML(tabulate.tabulate(table, headers = 'firstrow', tablefmt = 'html')))
```

### Modelo de redes neuronales

```python
from skfore.models.MLP import MLP
```

```python
warnings.filterwarnings('ignore')
model = MLP(p = 4, optim_type = 'auto')
model.fit(tse)
forecast = model.predict(ts = tse, periods = 20)
MSE_E = model.calc_error(ts = tse)
MAD_E = model.calc_error(ts = tse, error_function = sklearn.metrics.mean_absolute_error)
MSE_P = model.calc_error(ts = tsp)
MAD_P = model.calc_error(ts = tsp, error_function = sklearn.metrics.mean_absolute_error)
model.plot(tse, forecast)
```

```python
table = [["Error","Entrenamiento","Prueba"],
         ["MSE",MSE_E,MSE_P],
         ["MAD",MAD_E,MAD_P]]

display(HTML(tabulate.tabulate(table, headers = 'firstrow', tablefmt = 'html')))
```

### Modelo SARIMA

```python
from skfore.models.SARIMA import SARIMA
```

```python
model = SARIMA(p = 0,d = 1,q = 1,P = 0,D = 1,Q = 1,s = 12)
model.fit(tse)
forecast = model.predict(ts = tse, periods = 20)
MSE_E = model.calc_error(ts = tse)
MAD_E = model.calc_error(ts = tse, error_function = sklearn.metrics.mean_absolute_error)
MSE_P = model.calc_error(ts = tsp)
MAD_P = model.calc_error(ts = tsp, error_function = sklearn.metrics.mean_absolute_error)
model.plot(tse, forecast)
```

```python
table = [["Error","Entrenamiento","Prueba"],
         ["MSE",MSE_E,MSE_P],
         ["MAD",MAD_E,MAD_P]]

display(HTML(tabulate.tabulate(table, headers = 'firstrow', tablefmt = 'html')))
```
